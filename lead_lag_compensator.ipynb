```python
import numpy as np
import matplotlib.pyplot as plt
import control

# parameter for a lag compensator
a_lag = 30.0
b_lag = 60.0
a_lead = 10.0
b_lead = 5.0

# Simulate continous transfer function response
sys_lag = control.TransferFunction([a_lag, 1], [b_lag, 1])
sys_lead = control.TransferFunction([a_lead, 1], [b_lead, 1])
sys_lead_lag = control.TransferFunction([a_lag * a_lead, a_lag + a_lead, 1], 
                                        [b_lag * b_lead, b_lag + b_lead, 1])

plt.figure(0)
mag, phase, omega = control.bode_plot(sys_lag)
mag, phase, omega = control.bode_plot(sys_lead)
mag, phase, omega = control.bode_plot(sys_lead_lag)
```
